```python
def ternarySearch(l, r, key, ar):
    while r >= l:
         
        # Find mid1 and mid2
        mid1 = l + (r-l) // 3
        mid2 = r - (r-l) // 3
 
        # Check if key is at any mid
        if key == ar[mid1]:
            return mid1
        if key == mid2:
            return mid2
 
        # Since key is not present at mid,
        # Check in which region it is present
        # Then repeat the search operation in that region
        if key < ar[mid1]:
            # key lies between l and mid1
            r = mid1 - 1
        elif key > ar[mid2]:
            # key lies between mid2 and r
            l = mid2 + 1
        else:
            # key lies between mid1 and mid2
            l = mid1 + 1
            r = mid2 - 1
 
    # key not found
    return -1
```

Description
Sort a list of strings according to their length, with the smallest string appearing first.
Hint: Use the appropriate key in the sorted() function. If two strings are of the same length then their order  as in the original string  should be preserved.

Example:
Input 1:

Output 1:

```python
input_list = ['I', 'love', 'to', 'code', 'in', 'Python'] 
d1 = sorted(input_list, key = lambda s :len(s))
print(d1)
```
